```python
import tensorflow
import matplotlib.pyplot as plt
import seaborn as sns
import keras
import keras
```

### TRAINING DATASET

```python
training_data = keras.utils.image_dataset_from_directory(
    "train",
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
    pad_to_aspect_ratio=False,
    data_format=None,
    verbose=True,
)
```

```python
validation_data = keras.utils.image_dataset_from_directory(
    "valid",
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
    pad_to_aspect_ratio=False,
    data_format=None,
    verbose=True,
)
```

### BUILDING MODEL

# Sequential --> Conv2D --> MaxPool2D --> Flatten
https://stanford.edu/~shervine/teaching/cs-230/cheatsheet-convolutional-neural-networks

Sequnetial = Starting of the model
Conv2D = Adding the infomration of the file and

```python
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Dropout,Flatten
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Conv2D(filters= 32 , kernel_size= 3 , padding="same" , activation="relu" , input_shape = (128,128,3) ))

# Filter --> Number of filters we want
# kernal_size --> Size of the filter (3X3)
# input_shape --> Input shape of image ( image_size,image_size,3(rgb)) --> Passed only to first layer

model.add(Conv2D(filters= 32 , kernel_size= 3 , activation="relu" ))

model.add(MaxPool2D(pool_size= 2, strides= 2))

# Pool_size = This means the size of the window that moves across the image.
# Strides = This tells how far the window moves each time.
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Conv2D(filters= 64 , kernel_size= 3 , padding="same" , activation="relu" , input_shape = (128,128,3) ))
model.add(Conv2D(filters= 64 , kernel_size= 3 , activation="relu" ))
model.add(MaxPool2D(pool_size= 2, strides= 2))
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Conv2D(filters= 128 , kernel_size= 3 , padding="same" , activation="relu" , input_shape = (128,128,3) ))
model.add(Conv2D(filters= 128 , kernel_size= 3 , activation="relu" ))
model.add(MaxPool2D(pool_size= 2, strides= 2))
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
# Flattens the whole final layer we get
```

```python
model.add(Dense(units=1500, activation = "relu"))
```

```python
# Output Layer
model.add(Dense(units=38, activation= "softmax"))
```

```python
model.compile(optimizer=tensorflow.keras.optimizers.Adam(learning_rate=0.0001),loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
model.summary()
```

### MODEL TRAINING

```python
import tensorflow
from tensorflow.keras.callbacks import EarlyStopping

earlystopping = EarlyStopping(monitor="val_loss",patience=3,restore_best_weights=True)
```

```python
training_history = model.fit(training_data,validation_data=validation_data,epochs=10,callbacks=[earlystopping])
```

### EVALUATION OF MODEL

```python
train_loss,train_acc = model.evaluate(training_data)
print(f"Training Accuracy : {train_acc} and Training Loss : {train_loss}")
```

```python
val_loss,val_acc = model.evaluate(validation_data)
print(f"Validation Accuracy : {val_acc} , Validation Loss : {val_loss}")
```
